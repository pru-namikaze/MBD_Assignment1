```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
time_dataframe = dataframe
x = dataframe[['lights', 'T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility', 'Tdewpoint']]
y = dataframe['Appliances']
linear_regression = LinearRegression()
linear_regression.fit(x, y)
print('Intercept: \n', linear_regression.intercept_)
print('Coefficients: \n', linear_regression.coef_)

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
print_model = model.summary()
print(print_model)
```
